# Answer 1.1a

In the Non-Bayesian Latent Variable Model, we defined the ELBO as:

$$ ELBO(q, \theta, \phi) = \mathbb{E}_{q(z)}\big[ \log \frac{p(z, y|\theta, \phi)}{q(z)} \big] $$

In the Bayesian Latent Variable Model, introduce our our variational distribution $q(z, \theta, \phi)$, which implicitly has parameters (that I omit for brevity) and which factorizes as $q(z, \theta, \phi) = q(\theta|\lambda_{\theta}) q(\phi|\lambda_{\phi}) \prod_{n=1}^N q(z_n | \lambda_n)$ under the mean field assumption. We define the EBLO as follows:

\begin{align*}
ELBO(\lambda_{\theta}, \lambda_{\phi}, \lambda_{z}) 
&= \mathbb{E}_{q(z, \theta, \phi)}\big[ \log \frac{p(z, y, \theta, \phi |a, b)}{q(z, \theta, \phi)} \big]\\
&= \mathbb{E}_{q(z, \theta, \phi)}\big[ \log \frac{p(z, y, \theta, \phi |a, b)}{q(\theta|\lambda_{\theta}) q(\phi|\lambda_{\phi}) \prod_{n=1}^N q(z_n | \lambda_n)} \big]
\end{align*}

The two are similar in that we compare the log of the joint probability of all the variables (observed, latent and parameters) against the variational distribution's probability of the latents and parameters.
They differ in that the Bayesian Latent Variable Model has the parameters $\theta, \phi$ included in the variational distribution $q(\cdot)$.

# Answer 1.1b

In EM, we alternate between the E step, in which we set $q(z) = p(z|y, \theta)$, and the M-step, in which we optimize $\theta = \underset{\theta}{\mathrm{argmax}} \mathbb{E}_{q}[\log p(z, y| \theta)]$.

In Coordinate Ascent Variational Inference (CAVI), we typically can no longer solve explicitly for $q(z) = p(z|y, \theta)$ exactly and thus we update the parameters of our $q(\cdot)$ functions proportional to the exponential of the expectation. In the limiting case that $\theta, \phi$ are Dirac distributions/functions, the two models are equivalent.

# Answer 1.2a

Directly minimizing the KL-divergence in the original objective function is difficult because we don't know the parameters, so even if $p(z|y, \theta, \phi)$ has a nice analytical form and $q(z)$ has a similar form that together has a closed form solution (e.g. two multivariate Gaussians), we have no idea what the target $p(y|z, \theta, \phi)$ is because we don't know $\theta, \phi$.


# Answer 1.2b
In the case of the Gaussian Mixture Models in Lecture 7, the E step was made easy because the distribution $p(Z_n|Y_n, \mu_{\text{old}}, \sigma^2_{\text{old}}, \pi_{\text{old}})$ had a simple form (a Categorical distribution) and therefore finding a $q(Z_N)$ that minimized the KL divergence was easy - simply set the parameters equal to the parameters of $p(Z_n|Y_n, \mu, \sigma^2, \pi)$. # Answer 1.2c
In Lecture 8, we showed that maximizing $ELBO(\lambda)$, defined below,


\begin{align*}
\underset{\lambda_i}{\max} ELBO(\lambda) = \underset{\lambda_i}{\max} \mathbb{E}_{\psi \sim q(\psi|\lambda)}\left[\log\left(\frac{p(\psi, Y_1, \ldots, Y_N))}{q(\psi | \lambda)} \right)\right]
\end{align*}

is equivalent to minimizing the following KL divergence:

\begin{align*}
\underset{\lambda_i}{\min}D_{\text{KL}} \left[ q(\psi_{i}|\lambda_{i})\| \mathcal{Z}\exp\left\{\mathbb{E}_{\psi_{-i} \sim q(\psi_{-i}|\lambda_{-i})} \left[\log \left(p(Y_1, \ldots, Y_N, \psi_i | \psi_{-i}) \right)\right]\right\}\right].
\end{align*}

In this instance, maximizing the ELBO is difficult because the ELBO requires computing an expectation over samples from a distribution ($q(\psi_i | \lambda_i)$ while trying to optimize the parameters of said distribution ($\lambda_i$). Hence we prefer to minimize the KL divergence.


# Answer 1.2d
Based on this analysis, we would prefer to minimize the KL-divergence (1) when the conditional distribution over the latents has a simple form, or (2) when maximizing the ELBO requires changing parameters of the distribution we take the expectation with respect to.

3. **(The Mean Field Assumption and Coordinate Ascent)**  Describe exactly when and how the mean field assumption is used in the derivation of the coordinate ascent updates.

The mean field assumption states that we can factorize our joint latent variable distribution into the product of the marginals

$$q(z, \theta, \phi|\lambda_z, \lambda_{\theta}, \lambda_{\phi}) = q(z|\lambda_z) q(\theta | \lambda_{\theta}) q(\phi|\lambda_{\phi})$$

We use this assumption in the derivation of the coordinate ascent updates to separate $\psi_i$ from $\psi_{-i}$ in the step labeled Fubini's Theorem.

\begin{align*}
ELBO(\lambda)
 &= \int_{\Psi} \left[ \log \left(\frac{p(Y_1, \ldots, Y_N, \psi)}{q(\psi|\lambda)} \right)\right]q(\psi|\lambda) d\psi\\
&= \int_{\Psi_i} \int_{\Psi_{-i}} \left[ \log \left(\frac{p(Y_1, \ldots, Y_N, \psi_i | \psi_{-i})p(\psi_{-i})}{q(\psi_{i}|\lambda_{i})q(\psi_{-i}|\lambda_{-i})} \right)\right] q(\psi_{i}|\lambda_{i})q(\psi_{-i}|\lambda_{-i}) d\psi_{-i}d\psi_i\quad (\text{Fubini's Theorem})\\
\end{align*}

2. **(Empirical Bayes)** Using Empirical Bayes and moment matching, choose values for the hyperparameters $\alpha, \beta$ based on your data. Use these values of $\alpha$ and $\beta$ to obtain posterior distributions for each county.

***Hint:*** You'll first need to derive the fact that the ***evidence*** for a Poisson-Gamma model has a Negative Binomial distribution.

Starting with the hint, we derive the evidence for a Poisson-Gamma Model:

\begin{align}
p(y_j|\alpha, \beta)
&= \int_0^{\infty} p(y_j|\theta_j) P(\theta_j | \alpha, \beta) \, d\theta_j\\
&= \int_0^{\infty} \frac{(5 n_j \theta_j)^{y_j} e^{-5 n_j \theta_j}}{y_j !} \frac{\beta^{\alpha}}{\Gamma(\alpha)} \theta_j^{\alpha - 1} e^{-\beta \theta_j} \, d\theta_j\\
&= \frac{(5 n_j)^{y_j} \beta^{\alpha}}{y_j ! \Gamma(\alpha)} \int_0^{\infty} \theta_j^{y_j + \alpha - 1} e^{-\theta_j(5n + \beta)}\\
&= \frac{(5 n_j)^{y_j} \beta^{\alpha}}{y_j ! \Gamma(\alpha)} \frac{\Gamma(y_j + \alpha)}{(5n+\beta)^{y_j + \alpha}}\\
&= \frac{\Gamma(y_j + \alpha)}{y_j ! \Gamma(\alpha)} \frac{(5 n_j)^{y_j} \beta^{\alpha}}{(5n+\beta)^{y_j + \alpha}}\\
&= \frac{\Gamma(y_j + \alpha)}{y_j ! \Gamma(\alpha)} \Big(\frac{5 n_j }{5 n_j + \beta} \Big)^{y_j}  \Big( \frac{\beta }{5 n_j + \beta} \Big)^{\alpha}\\
&= \text{Negative Binomial}(\alpha, \frac{5 n_j}{5 n_j + \beta})
\end{align}

Using the property that the mean of a Negative Binomial distribution is $pr/(1-p)$ and the variance is $pr/(1-p)^2$, we find that

\begin{align*}
\mathbb{E}[\frac{y_j}{n_j}]
&= \frac{1}{n_j}\frac{5 n_j}{5 n_j + \beta} \alpha \frac{5 n_j +\beta}{\beta}\\
&= \frac{5 \alpha}{\beta}\\
\mathbb{V}[\frac{y_j}{n_j}]
&= \frac{1}{\bar{n}} \frac{5 }{5 \bar{n} + \beta} \alpha (\frac{5 \bar{n} +\beta}{\beta})^2\\
&= \frac{5 \alpha (5 \bar{n} + \beta)}{\bar{n} \beta^2}
\end{align*}

Solving for the parameters $\alpha, \beta$:

\begin{align*}
    \mu = \frac{5 \alpha}{\beta} &\Rightarrow \beta = \frac{5 \alpha}{\mu}\\
    \sigma^2 = \frac{5 \alpha (5 \bar{n} + \beta)}{\bar{n} \beta^2}
    &\Rightarrow \alpha = \frac{\mu^2 \bar{n}}{\bar{n}\sigma^2 - \mu}\\
    \beta = \frac{5 \alpha}{\mu} &\Rightarrow \beta = \frac{5 \bar{n} \mu }{\bar{n} \sigma^2 - \mu}
\end{align*}

```python
pct_mortality = kcancer['pct_mortality'].values
mu = np.mean(pct_mortality)
sigma_sqed = np.var(pct_mortality)
n_bar = np.mean(kcancer['pop'].values)

alpha = np.square(mu)*n_bar / (n_bar*sigma_sqed - mu)
beta = 5 * alpha / mu
print('Beta: ', beta)
print('Alpha: ', alpha)
```

## Part II: Hierarchical Bayes
Rather than choosing fixed constants for the hyperparameters $\alpha, \beta$, following the Bayesian philosophy, we typically put additional priors on quantities of which we are uncertain. That is, we model the kidney cancer rates using a ***hierarchical model***:

\begin{aligned}
y_j| \theta_j &\sim Poisson(5 \cdot n_j \cdot \theta_j), \quad j = 1, \ldots, N\\
\theta_j | \alpha, \beta &\sim Ga(\alpha, \beta), \quad j = 1, \ldots, N\\
\alpha &\sim Ga(a, b)\\
\beta &\sim Ga(c, d)
\end{aligned}
where $a, b, c, d$ are hyperparameters.

Highlight the bottom 300 raw cancer rates in blue. Finally, on the same plot again, scatter plot the mean of the posterior marginal distribution over $\theta_j$, i.e. $p(\theta_j|y_1, \ldots, y_N)$, vs the county population size, highlight these means in orange. 

  You should use your sampler from the previous homework to obtain these posterior marginal samples.<br><br>

```python
# copied from HW3 solutions

#function for sampling a pair (alpha, beta) using Metropolis-Hastings
def draw_alpha_beta(alpha, beta, y, theta, steps):
    #initialization 
    old_alpha = alpha
    old_beta = beta
    accept = 0
    
    for step in range(steps):
        #propose a pair (alpha, beta)
        new_alpha, new_beta = np.random.multivariate_normal([old_alpha, old_beta], [[sigma_alpha, sigma_cross], [sigma_cross, sigma_beta]], 1)[0]
        #compute the log of the ratio of pdf's evaluated at the new and old alpha, beta values
        log_accept_frac = p_alpha_beta(new_alpha, new_beta, theta) - p_alpha_beta(old_alpha, old_beta, theta)
        #Accept mechanism
        if np.log(np.random.uniform()) < log_accept_frac:
            old_alpha = new_alpha
            old_beta = new_beta
            accept += 1
            
    return old_alpha, old_beta, accept * 1. / steps
    
#gibbs Sampling
def gibbs(y, n, alpha_init, beta_init, theta_init, iters, burn, thin=1):   
    #initialization 
    alphas = [alpha_init]
    betas = [beta_init]
    theta = theta_init
    accept_alpha_beta = 0
    samples = [theta]
    
    
    for i in range(iters):
            
        #draw alpha, beta
        alpha, beta, accept = draw_alpha_beta(alphas[-1], betas[-1], y, theta, 1)
        alphas.append(alpha)
        betas.append(beta)
        
        accept_alpha_beta += accept
        
        if (i % 5000 == 0) and i > 0:
            print('iteration {}, accept rate for MH {}'.format(i, accept_alpha_beta * 1. / i))

        #draw theta
        theta = draw_theta(y, n, alphas[-1], betas[-1])
        samples.append(theta)
        
    return np.array(samples[int(burn * iters)::thin]), accept_alpha_beta / iters, alphas[int(burn * iters)::thin], betas[int(burn * iters)::thin]
```

```python
# copied from HW3 solutions

df = pd.read_csv('kcancer.csv')

#set hyperparameters
a = 9
b = 6
c = 9
d = 0.00001
sigma_alpha = 0.1
sigma_cross = 1
sigma_beta = 1000

#set observed variables
N = len(df)
n = df['pop'].values
y = df['dc'].values

#Function for sampling a theta for a particular county
draw_theta = lambda y_j, n_j, alpha, beta: np.random.gamma(y_j + alpha, scale=1. / (5. *  n_j + beta))

#The joint pdf of alpha and beta
p_alpha_beta = lambda alpha, beta, theta: np.log(sp.stats.gamma.pdf(alpha, a, scale=1. / b)) \
                                          + np.log(sp.stats.gamma.pdf(beta, c, scale=1. / d)) \
                                          + alpha * np.log(theta).sum() - beta * theta.sum() \
                                          + N * alpha * np.log(beta) \
                                          - N * np.log(sp.special.gamma(alpha))                 
```

```python
#Initialize all parameters
alpha_init = 1.5451734918795321
beta_init = 133514.17193888038
theta_init = draw_theta(y, n, alpha_init, beta_init)

#Sample 100,000 samples with burn-in of 20% and thinning of 2
iters = 100001
thin = 20
burn = 0.2
samples, accept_rate, alphas, betas = gibbs(y, n, alpha_init, beta_init, theta_init, iters, burn, thin=thin)
print('\n\nAccept rate of M-H step:', accept_rate)
```

```python
from IPython.display import Image
import plotly
import plotly.graph_objs as go
import plotly.io as pio


x = kcancer['pop'].values
y = kcancer['pct_mortality'].values
sorted_indices = np.argsort(y)

x_hier_bayes = x
y_hier_bayes = np.mean(samples, axis=0)

x_low = x[sorted_indices[:300]]
y_low = y[sorted_indices[:300]]

x_med = x[sorted_indices[300:-300]]
y_med = y[sorted_indices[300:-300]]

x_high = x[sorted_indices[-300:]]
y_high = y[sorted_indices[-300:]]


data = [
    go.Scatter(x=np.log(x_emp_bayes), y=5*y_emp_bayes, name='Empirical Bayes Posterior Mean', mode='markers', marker_color='rgba(0, 0, 0, .1)'),
    go.Scatter(x=np.log(x_hier_bayes), y=5*y_hier_bayes, name='Hierarchical Posterior Mean', mode='markers', marker_color='rgba(0, 255, 50, .1)'),
    go.Scatter(x=np.log(x_low), y=y_low, name='Low', mode='markers', marker_color='rgba(0, 0, 255, .1)'),
    go.Scatter(x=np.log(x_high), y=y_high, name='High', mode='markers', marker_color='rgba(255, 0, 0, .1)'),
]

layout = dict(
    title='Cancer Rate by County Population',
    yaxis=dict(title='Cancer Rate'),
    xaxis=dict(title='Log County Population'))

fig = go.Figure(data=data, layout=layout)
# fig.show()
Image(pio.to_image(fig, format='png'))
```
