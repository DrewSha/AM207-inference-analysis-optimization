# Review of Hidden Markov Models and State Space Models

## Hidden Markov Models

The problem with types of sequential data observed from real-life dynamic system is that the observation is usually noisy. What we observe is not the true state of the system (e.g. the true state of the patient or the true location of the robot), but some signal that is corrupted by environmental noise. 

In a ***hidden Markov model (HMM)***, we assume that we do not have access to the values in the underlying Markov process
$$
X_{n+1} | X_{n} \sim p(X_{n+1} | X_{n}) \quad \mathbf{(state\; model)}
$$

and, instead, observe the process 
$$
Y_n | X_n \sim p(Y_n | X_n) \quad \mathbf{(observation\; model)}
$$
where $p(Y_n | X_n)$ is the probability density associated with observing $Y_n \in \mathcal{Y}$ given the latent value, $X_n$, of the underlying Markov process at time $n$.

<img src="fig/hmm.png" style="height:150px;" align="center">

If the state space is continuous, Hidden Markov Models are often called ***State-Space Models***.

## Example: Linear Gaussian Models

Let our state and observation spaces be Euclidean, $\mathcal{X} = \mathbb{R}^M$ and $\mathcal{Y} = \mathbb{R}^{M'}$. The transitions and observations in a ***linear Gaussian model*** are defined by linear transformations of Gaussian variables with the addition of Gaussian noise:

\begin{align}
&X_0 \sim \mathcal{N}(0, \Sigma)\\
&X_{n+1} = AX_n + B + C\xi \quad \mathbf{(State\;Model)}\\
&Y_{n+1} = DX_{n+1} + E + F\epsilon \quad \mathbf{(Observation\;Model)}
\end{align}

where $\xi \sim \mathcal{N}(0, \mathbf{I}_M)$, $\epsilon \sim \mathcal{N}(0, \mathbf{I}_{M'})$. Thus, the transitions and observations probability densities are:

\begin{align}
X_{n+1}|X_n &\sim \mathcal{N}(AX_n + B, CC^\top)\\
Y_{n+1} | X_{n+1} &\sim \mathcal{N}(DX_{n+1} + E, FF^\top)
\end{align}

Linear Gaussian Models are widely used in target tracking and signal processing, since inference for these models are tractible (we only need to manipulate Gaussians).

## Learning and Inference for HMMs
There are a number of inference problems associated with HMMs: <br>
**1. (Learning)** learning the dynamics of the state or observation model <br>
**2. (Inference)** estimating the probability distribution over one or more of the latent variables, $\{X_n\}_{n\geq 1}$, given a sequence of observations, $\{Y_n\}_{n\geq 1}$.

Inference, the task of estimating the latent variables, can be further broken down:

**1. Filtering** computing the distribution of the last latent variable in the sequence given all observations, $p(X_n| Y_{1:n})$.<br><br>
**2. Smoothing** computing the distribution of a hidden latent variable in a past time step given all observations, $p(X_t| Y_{1:n})$ where $t < n$.<br><br>
**3. Most probable explanation** computing the joint distribution of all latent variables given all observations, $p(X_{1:n}| Y_{1:n})$. Or, alternatively, compute the most likely sequence of latent variable values given all observations:
$$
x^*_{1:n} = \underset{x_{1:n}}{\mathrm{argmax}}\; p(x_{1:n} | y_{1:n})
$$

## Inducitive Algorithm for Inference

Notice that given $p(X_{n-1} | Y_{1:n-1})$, we can compute $p(X_n| Y_{1:n}) = \frac{p(Y_n | X_n) p(X_n | Y_{1:n-1})}{p(Y_n | Y_{1:n-1})}$. This leads to the following inducitive algorithm for infering $p(X_n| Y_{1:n})$: 

0. (**Inductive Hypothesis**) suppose we have $p(X_{n-1} | Y_{1:n-1})$
1. (**Prediction step**) compute $p(X_n | Y_{1:n-1}) = \int p(X_n|X_{n-1})p(X_{n-1}|Y_{1:n-1})dX_{n-1}$ 
2. (**Update step**) compute $p(X_n| Y_{1:n}) =\frac{p(Y_n|X_n)p(X_n|Y_{1:n-1})}{p(Y_n | Y_{1:n-1})}$, where $$p(Y_n | Y_{1:n-1}) = \int p(X_{n-1} | Y_{1:n-1}) p(X_n | X_{n-1}) p(Y_n | X_n) dX_{n-1:n}.$$

**The problem:** save for in very few cases, the integrals we need to compute in both the prediction and update steps are intractable! 

In the special case of **linear Gaussian models**, the distributions in the prediction and update steps are Gaussian and can be computed in closed form, the resulting iterative filtering algorithm is known as the ***Kalman Filter***.

## Limitations of Kalman Filters

When the observations are generated from a linear Gaussian system (LGS) and the dynamics of that system are exactly estimated then Kalman filters are guaranteed to provide estimates $\hat{x}_n$ of the latent state $x_n$ that minimizes mean squared error. This is why these filters are often called ***optimal***. 

1. **(Non-linearity)** As you've seen from HW#9, when we apply Kalman filters to observations that are not generated by an LGS, our estimate $\hat{x}_n$ of the latent state $x_n$ can be arbitrarily wrong -- the filter ***diverges***.<br><br>

2. **(Numerical errors)** Even on observations from an LGS, in practice, Kalman filters can suffer from filter divergence. This is typically causes by inaccurate estimates of the dynamics, as well as due to numerical roundoff errors. There are a number of variants of the Kalman filter that makes it robust to these erroros.

# Non-linear Dynamics

## Example: Non-linear Gaussian Models

Let our state and observation spaces be Euclidean, $\mathcal{X} = \mathbb{R}^M$ and $\mathcal{Y} = \mathbb{R}^{M'}$. The transitions and observations in a non-linear Gaussian model are defined by non-linear transformations of Gaussian variables with the addition of Gaussian noise:

\begin{align}
&X_0 \sim \mathcal{N}(0, \Sigma)\\
&X_{n+1} = f_\theta(X_n) + C\xi \quad \mathbf{(State\;Model)}\\
&Y_{n+1} = g_\phi(X_{n+1}) + F\epsilon \quad \mathbf{(Observation\;Model)}
\end{align}

where $f_\theta$, $g_\phi$ are nonlinear functions, $\xi \sim \mathcal{N}(0, \mathbf{I}_M)$, $\epsilon \sim \mathcal{N}(0, \mathbf{I}_{M'})$.

## Extended Kalman Filters
The main idea for handling non-linear Gaussian models: approximate the dynamics with linear dynamics and apply Kalman filter using the approximate linear dynamics:

\begin{align}
&X_{n+1} \approx f_\theta(\hat{x}_n) + \nabla f_\theta\vert_{\hat{x}_n}^\top(X_{n} - \hat{x}_n) + C\xi \quad \mathbf{(Approx.\; State\;Model)}\\
&Y_{n+1} \approx g_\phi(m_n) + \nabla g_\phi\vert_{m_n}^\top(X_{n+1} - m_n) + F\epsilon \quad \mathbf{(Approx.\;Observation\;Model)}
\end{align}

where $m_n$ is the mean of $p(X_{n+1}:Y_{1:n})$ computed in the prediction step of the Kalman filter.

Now we can compute the two steps in the Kalman filter algorithm using the linearized dynamics. This algorithm is called the ***extended Kalman filter***.

## Limitations of Extended Kalman Filters

Extended Kalman filters (EKFs) work well when the linearized dynamics are good at approximating the underlying non-linear dynamics. 

Unfortunately, many functions are not well approximated by their first order Taylor expansions. In these cases EKFs can diverge -- the estimated $\hat{x}_n$ can be very far off from the true value of $x_n$.

Numerical error issues that affect Kalman filters also applies here.

# Sequential Monte Carlo

## Importance Sampling as a Sampling Method
We've seen importance sampling as Monte Carlo method. It turns out that one can turn importance sampling into a sampler by representing the target distribution, say $p(X_{0:n} | Y_{1:n})$, as a set of importance weighted samples:
$$
p(X_{0:n} | Y_{1:n}) \approx \sum_{s=1}^S W_n^s \delta_{X_{0:n}^s}(X_{0:n})
$$
where $X_{0:n}^s$ are samples drawn from some proposal importance distribution $q_n(X_{0:n})$, $\delta_{X_{0:n}^s}$ is the dirac distribution, $W_n^s$ are importance weights computed to adjust the empirical distribution of samples from $q_n$ to appear like samples from $p$. The weights $W_n^s$ are computed as:

$$
W_n^s = \frac{w_n^s}{Z_n},
$$
where $w_n^s = \frac{p(X_{0:n}, Y_{0:n})}{q_n(X_{0:n})}$ is the un-normalized importance weight and $Z_n$ is the normalizing constant
$$
Z_n = \frac{1}{S}\sum_{s=1}^{S} w_n^s.
$$

## The Idea of Sequential Monte Carlo

Rather than computing the distribution $p(X_n|Y_{1:n})$ for each $n$ analytically as we did for Kalman filters, we approximate each distribution empirically with $S$ number of samples obtained by importance sampling (as a sampler). Estimate $\hat{x}_n$ of the latent state $x_n$ are computed as the mean of these samples.
<img src='fig/sis.png' style='width:600px'>

## Sequential Importance Sampling

The problem with importance sampling is that in order to compute the weight $w_n^s = \frac{p(X_{0:n} | Y_{1:n})}{q(X_{0:n})}$, we need to know $p(X_{0:n} | Y_{1:n})$ -- this means an expensive recomputation when a new observation arrives! It turns out if we use a structured importance distribution $q$, we can compute $w_n^s$ inductively.

Assume the following factorization for $q$ for every $n$:
$$
q(X_{0:n}) = q(X_{0:n-1})q(X_n | X_{0:n-1})
$$
We can rewrite the importance weight as:

\begin{align}
w_n^s &= \frac{p(X_{0:n} | Y_{1:n})}{q(X_{0:n})}\\
&\propto \frac{p(Y_n|X^s_n)p(X^s_n|X^s_{n-1})p(X^s_{0:n-1} | Y_{0:n-1})}{q(X_{0:n})}\\
&= \frac{p(Y_n|X^s_n)p(X^s_n|X^s_{n-1})p(X^s_{0:n-1} | Y_{0:n-1})}{q(X_{0:n-1})q(X_n | X_{0:n-1})}\\
&= \frac{p(Y_n|X^s_n)p(X^s_n|X^s_{n-1})}{q(X_n | X_{0:n-1})} \underbrace{\frac{p(X^s_{0:n-1} | Y_{0:n-1})}{q(X_{0:n-1})}}_{w_{n-1}^s}
\end{align}
We get an inductive way to update the importance weights:
$$
w_n^s \propto \frac{p(Y_n|X_n^{s})p(X_n^{s}|X_{n-1}^{s})}{q(X_{n}^{s}|X_{0:n-1}^{s})}w_{n-1}^s
$$

## The Sequential Importance Sampling Algorithm

The incremental computation of the importance weights can be summarized in the ***sequential importance sampling (SIS)*** algorithm, which is an example of a ***sequential Monte Carlo*** or ***particle filter*** method:

**Initialize**: $t = 0$<br>
**1.** For $s = 1,...,S$:<br>
  $\quad$ Sample $X_0^{s}\sim p(X_0)$<br>
  $\quad$ Set $w_0^{s}=p(X_0^{s})$
  
**Iterate**: for $n$ in $1\ldots N$<br>
**1.** For $s = 1,...,S$:<br>
  $\quad$ Sample $X_n^{s}\sim q(X_n|X_{n-1}^{s})$ and set $X_{0:n}^{s}=(X_{n-1}^{s},X_n^{s})$<br>
    
  $\quad$ Set $w_n^{s}\propto\frac{p(Y_n|X_n^{s})p(X_n^{s}|X_{n-1}^{s})}{q(X_{n}^{s}|X_{0:n-1}^{s})}w_{n-1}^{s}$
  
**2.** For s = 1,...,S normalize $w_n^{s} = \frac{w_n^{s}}{\sum_{s'}w_n^{s'}}$

## Limitations of Sequential Monte Carlo

The uncertainty of the estimate $\hat{x}_n$ in SIS tends to collapse as $n$ becomes large. This is because the importance weights become more and more heterogenous, where the weight of most samples are zero and a few samples have large weights.

This tendency for the weights to become heterogeneous can be seen from the weight update

$$
w_n^s \propto \frac{p(Y_n|X_n^{s})p(X_n^{s}|X_{n-1}^{s})}{q(X_{n}^{s}|X_{0:n-1}^{s})}w_{n-1}^s
$$

Once $w_{n-1}^s$ becomes small, then, generally speaking, $w_n^s$ becomes smaller and smaller as $n$ increases.

Two common methods to prevent ***degeneracy*** in the importance weights are: (1) prune samples with small weights (selection) and (2) resample from the weighted samples after pruning (resampling).

```python
#function for simulating observations from a stochastic volatility model
def SVM_simulate(sigma, alpha, beta, sample_x, steps=100):
    x_current = sample_x(1)
    latent_vars = [x_current]
    observations = []
    
    for _ in range(steps):
        y_current = beta * np.exp(x_current / 2.) * np.random.normal(0, 1) 
        x_next = alpha * x_current + sigma * np.random.normal(0, 1)
        observations.append(y_current)
        latent_vars.append(x_next)
        x_current = x_next
        
    return np.array(latent_vars[:-1]), np.array(observations)
```

```python
#function for simulating observations from a linear gaussian model
def LGM_simulate(a, b, c, d, e, f, sample_x, steps=100):
    x_current = sample_x(1)
    latent_vars = [x_current]
    observations = []
    
    for _ in range(steps):
        y_current = d * x_current + e + f * np.random.normal(0, 1)
        x_next = a * x_current + b  + c * np.random.normal(0, 1)
        observations.append(y_current)
        latent_vars.append(x_next)
        x_current = x_next
        
    return np.array(latent_vars[:-1]), np.array(observations)
```

```python
#implementation of Kalman filtering
def kalman_filter(y, a, b, c, d, e, f, mu_init=0, sigma_init=1.):
    steps = len(y)
    x_mean_current = mu_init
    x_sigma_current = sigma_init
    x_mus = [mu_init]
    x_sigmas = [sigma_init]

    for n in range(1, steps):
        #predict step
        mu = a * x_mean_current + b 
        r = a**2 * x_sigma_current + c**2
        y_current = d * mu + e
        
        #update step
        K = r * d / (f**2 + d**2 * r)
        x_mean_next = mu + K * (y[n] - y_current)
        x_sigma_next = r * (1 - K * d)
        
        #prep for next iteration
        x_mus.append(x_mean_next[0])
        x_sigmas.append(x_sigma_next)
        
        x_mean_current = x_mean_next
        x_sigma_current = x_sigma_next
        
    return np.array(x_mus), np.array(x_sigmas)
```

```python
#implementation of sequential importance sampling
def SIS(y, q_init_sample, q_init_eval, q_sample, q_eval, S=50):
    #base case: time index 0
    
    x_current = q_init_sample(S)
    x_samples = [x_current]
    log_weights = np.log(q_init_eval(x_current))
    weights_current = log_weights - logsumexp(log_weights)
    x_log_weights = [weights_current]
    
    
    #inductive case: time index n
    for n in range(1, steps):
        x_next = q_sample(x_current, 1)
        log_weights = weights_current + np.log(q_eval(y[n], x_next))
        weights_next = log_weights - logsumexp(log_weights)
        
        x_samples.append(x_next)
        x_current = x_next
        
        x_log_weights.append(weights_next)
        weights_current = weights_next
    
    x_samples = np.array(x_samples)
    x_log_weights = np.array(x_log_weights)
    x_means = np.sum(np.multiply(x_samples, np.exp(x_log_weights)), axis=1)
    x_sigmas = np.sum(np.multiply((x_samples - x_means[:, np.newaxis])**2, 
                                  np.exp(x_log_weights)), axis=1)
    
    return x_means, x_sigmas
```

```python
###define the parameters of the Linear Gaussian Model

#parameters for the underlying markov process: x_n+1 = a * x_n + b + c * N(0, 1)
a = 1.01
b = 0.2
c = 1.

#parameters for the observation model: y_n = d * x_n + e + f * N(0, 1)
d = 1.
e = 0.
f = np.sqrt(10)

#number of time steps to simulate
steps = 100

#initial distribution over latent state space
sample_x = lambda n: np.random.normal(0, 1, size=n)

###generate data by simulating the LGM
latent_vars, observations = LGM_simulate(a, b, c, d, e, f, sample_x, steps=steps)
```

```python
###estimate latent states given observations using Kalman filtering
x_mus, x_sigmas = kalman_filter(observations, a, b, c, d, e, f, mu_init=0, sigma_init=10)

###estimate latent states given observations using SIS
q_init_sample = lambda n: np.random.normal(0, 1, n)
q_init_eval = lambda x: (2 * np.pi)**-0.5 * np.exp(-0.5 * x**2)

def q_sample (x, n): 
    return np.array([np.random.normal(loc=a * mu + b, scale=c, size=n) for mu in x]).flatten()

q_eval = lambda y, x: (2 * np.pi * f**2)**-0.5 * np.exp(-0.5 * (y - (d * x + e))**2 / f**2)
x_means_sis, x_sigmas_sis = SIS(observations, q_init_sample, q_init_eval, q_sample, q_eval, S=100)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))

###compare true latent states with estimated latent states from Kalman filtering

#plot ground truth
ax[0].plot(np.arange(steps), latent_vars, 
        color='red', label='latent states')
ax[0].scatter(np.arange(len(observations)), observations,
           color='black', s=5., alpha=0.3, label='observations')

#plot estimated mean of marginal posterior over latent states
ax[0].plot(np.arange(steps), x_mus, 
        color='blue', label='Kalman estimate mean over latent states')

#plot estimated variance of marginal posterior over latent states
ax[0].fill_between(np.arange(steps), x_mus + 2 * x_sigmas, x_mus - 2*x_sigmas, 
                color='blue', alpha=0.1, 
                label='Kalman estimate variance over latent states, +/- 2 std')

ax[0].legend(loc='best')
ax[0].set_xlabel('Time Index')
ax[0].set_title('Kalman Filter Estimate of LGM States')

###compare true latent states with estimated latent states from SIS

ax[1].plot(np.arange(steps), latent_vars, 
        color='red', label='latent states')
ax[1].scatter(np.arange(len(observations)), observations,
           color='black', s=5., alpha=0.3, label='observations')

#plot estimated mean of marginal posterior over latent states
ax[1].plot(np.arange(steps), x_means_sis, 
        color='blue', label='SIS estimate mean over latent states')

#plot estimated variance of marginal posterior over latent states
ax[1].fill_between(np.arange(steps), x_means_sis + 2 * x_sigmas_sis, x_means_sis - 2*x_sigmas_sis, 
                color='blue', alpha=0.1, 
                label='SIS estimate variance over latent states, +/- 2 std')

ax[1].legend(loc='best')
ax[1].set_xlabel('Time Index')
ax[1].set_title('SIS Estimate of LGM States')
plt.show()
```

```python
###define the parameters of the Stochastic Volatility Model

#parameters for the underlying markov process: x_n+1 = alpha * x_n + sigma * N(0, 1)
alpha = 0.91
sigma = 1.

#parameters for the observation model: y_n = beta * exp(x_n/2) * N(0, 1)
beta = 0.5

#number of time steps to simulate
steps = 100

#initial distribution over latent state space
sample_x = lambda n: np.random.normal(0, sigma**2 / (1. - alpha**2), size=n)

###generate data by simulating the SVM
latent_vars, observations = SVM_simulate(sigma, alpha, beta, sample_x, steps=steps)
```

```python
init_std = sigma**2 / (1. - alpha**2)

###estimate latent states given observations using Kalman filtering
x_mus, x_sigmas = kalman_filter(observations, alpha, 0., sigma, 1., 0., beta, mu_init=0, sigma_init=init_std)

###estimate latent states given observations using SIS
q_init_sample = lambda n: np.random.normal(0, init_std, n)
q_init_eval = lambda x: (2 * np.pi * init_std)**-0.5 * np.exp(-0.5 * x**2 / init_std)

def q_sample (x, n): 
    return np.array([np.random.normal(loc=alpha * mu, scale=sigma, size=n) for mu in x]).flatten()

q_eval = lambda y, x: (2 * np.pi * beta**2 * np.exp(x))**-0.5 * np.exp(-0.5 * y**2 / (beta**2 * np.exp(x)))

x_means_sis, x_sigmas_sis = SIS(observations, q_init_sample, q_init_eval, q_sample, q_eval, S=100)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))

###compare true latent states with estimated latent states from Kalman filtering

#plot ground truth
ax[0].plot(np.arange(steps), latent_vars, 
        color='red', label='latent states')
ax[0].scatter(np.arange(len(observations)), observations,
           color='black', s=5., alpha=0.3, label='observations')

#plot estimated mean of marginal posterior over latent states
ax[0].plot(np.arange(steps), x_mus, 
        color='blue', label='Kalman estimate mean over latent states')

#plot estimated variance of marginal posterior over latent states
ax[0].fill_between(np.arange(steps), x_mus + 2 * x_sigmas, x_mus - 2*x_sigmas, 
                color='blue', alpha=0.1, 
                label='Kalman estimate variance over latent states, +/- 2 std')

ax[0].legend(loc='best')
ax[0].set_xlabel('Time Index')
ax[0].set_title('Kalman Filter Estimate of SVM States')

###compare true latent states with estimated latent states from SIS

ax[1].plot(np.arange(steps), latent_vars, 
        color='red', label='latent states')
ax[1].scatter(np.arange(len(observations)), observations,
           color='black', s=5., alpha=0.3, label='observations')

#plot estimated mean of marginal posterior over latent states
ax[1].plot(np.arange(steps), x_means_sis, 
        color='blue', label='SIS estimate mean over latent states')

#plot estimated variance of marginal posterior over latent states
ax[1].fill_between(np.arange(steps), x_means_sis + 2 * x_sigmas_sis, x_means_sis - 2*x_sigmas_sis, 
                color='blue', alpha=0.1, 
                label='SIS estimate variance over latent states, +/- 2 std')

ax[1].legend(loc='best')
ax[1].set_xlabel('Time Index')
ax[1].set_title('SIS Estimate of SVM States')
plt.show()
```
