# Review of Variational Autoencoders

## Variational Autoencoders

A ***variational autoencdoer*** is both a latent variable model and an inference method. In VAEs, we assume that the data is generated by the following process:
<table>
    <tr>
        <td>
            <img src="fig/VAE_graphical.png" style="height:200px;" align="center">
        </td>
        <td>
            \begin{align}
            z_n &\sim p_\theta(z), \; \mathbf{(prior)}\\
            y_n &\sim p_\theta(y_n|z_n), \; \mathbf{(likelihood)}
            \end{align}
        </td>
    </tr>
</table>
where the likelihood $p_\theta(y_n|z_n)$ is defined by $y_n = f_\theta(z_n) + \epsilon$, $\epsilon \sim p(\epsilon)$. The goal of a VAE model is to:

$\mathbf{1.}$ learn the function $f_\theta$, ***the generative network***, to mimic the empirical distribution of observed data <br><br>
$\mathbf{2.}$ learn a function $g_\phi$, ***the inference network***, to infer the best Gaussian approximation $\mathcal{N}(\mu(y_n), \Sigma(y_n))$ to the posterior $p(z_n|y_n)$ given an observation $y_n$

## Inference in Variational Autoencoders

The generative network $f_\theta$ and inference networks $g_\phi$ are trained jointly by gradient descent to maximize the ELBO
$$
\theta^*, \phi^* = \mathrm{argmax}_{\theta, \phi} \sum_{n=1}^N \mathbb{E}_{z\sim q_{\phi}(z_n)}\left[\log\frac{p_\theta(y_n, z_n)}{q_{\phi}(z_n)} \right]
$$
where $q_{\phi}(z_n) = \mathcal{N}(z_n; \mu(y_n), \Sigma(y_n))$. You can interpret the joint training of $f_\theta$ and $g_\phi$ to be performing both the E and the M steps of EM for the latent variable model.
<img src="./fig/VAE.jpg" style="height: 280px;" align="center"/>

## How to Evaluate a Trained VAE Model
We frequently compute the likelihood of the observed data $\prod_{n=1}^N p_\theta(y_n)$ under the learned $\theta$. There are, furthermore, image and text specific metrics for measuring the quality of the generative model. Finally, we frequently sample from $p_\theta(z)$ and visualize samples from $y\sim p_\theta(y|z)$:

# Hidden Markov Models and State Space Models

## Markov Models

Recall that a ***discrete time stochastic process*** is a sequence of random variables $\{X_n\}_{n\geq 0}$ that takes values in the space $\mathcal{X}$, this is called ***the state space***. The state space can be continuous or discrete, finite or infinite. A stochastic process is a ***discrete time Markov model*** if the process is given by

$$
\begin{align}
X_1 &\sim p(X_1)\\
X_{n+1} | X_{n} &\sim p(X_{n+1} | X_{n})
\end{align}
$$

where $p(X_{n+1} | X_{n})$ is the probability density associated with transitioning from one time step to the next and $p(X_1)$ is the probability density associated with the initial time step.

<img src="fig/markov.png" style="height:80px;" align="center">

## Markov Models and Sequential Data

We have used Markov models to model the samples obtained by a MCMC sampler over iterations. But Markov models can be used more generally to model any data of a ***sequential*** nature -- data where the ordering of the observation contains information.

Sequential data appears in many domains:
1. biometric readings of patients in a hospital over their stay
2. gps location of autonomous machines over time
3. economic/financial indicators of a system over time

Learning how Markov model evolves over time, the ***dynamics***, can lend insights on the mechanics of real-life systems that the model represent.

## Hidden Markov Models

The problem with types of sequential data observed from real-life dynamic system is that the observation is usually noisy. What we observe is not the true state of the system (e.g. the true state of the patient or the true location of the robot), but some signal that is corrupted by environmental noise. 

In a ***hidden Markov model (HMM)***, we assume that we do not have access to the values in the underlying Markov process
$$
X_{n+1} | X_{n} \sim p(X_{n+1} | X_{n}) \quad \mathbf{(state\; model)}
$$

and, instead, observe the process 
$$
Y_n | X_n \sim p(Y_n | X_n) \quad \mathbf{(observation\; model)}
$$
where $p(Y_n | X_n)$ is the probability density associated with observing $Y_n \in \mathcal{Y}$ given the latent value, $X_n$, of the underlying Markov process at time $n$.

<img src="fig/hmm.png" style="height:150px;" align="center">

If the state space is continuous, Hidden Markov Models are often called ***State-Space Models***.

#### Notation:

We will observe the following convetions:

- A random variable in the HMM at time index $n$ is denoted $X_n$ or $Y_n$.
- A collection of random variables from time index $n$ to time index $n + k$ is denoted $X_{n:n+k}$ or $Y_{n:n+k}$.
- The value of the random variable at time index $n$ is denoted $x_n$ or $y_n$.

## Linear Gaussian Models

Let our state and observation spaces be Euclidean, $\mathcal{X} = \mathbb{R}^M$ and $\mathcal{Y} = \mathbb{R}^{M'}$. The transitions and observations in a ***linear Gaussian model*** are defined by linear transformations of Gaussian variables with the addition of Gaussian noise:

\begin{align}
&X_0 \sim \mathcal{N}(0, \Sigma)\\
&X_{n+1} = AX_n + B + C\xi \quad \mathbf{(State\;Model)}\\
&Y_{n+1} = DX_{n+1} + E + F\epsilon \quad \mathbf{(Observation\;Model)}
\end{align}

where $\xi \sim \mathcal{N}(0, \mathbf{I}_M)$, $\epsilon \sim \mathcal{N}(0, \mathbf{I}_{M'})$. Thus, the transitions and observations probability densities are:

\begin{align}
X_{n+1}|X_n &\sim \mathcal{N}(AX_n + B, CC^\top)\\
Y_{n+1} | X_{n+1} &\sim \mathcal{N}(DX_{n+1} + E, FF^\top)
\end{align}

Linear Gaussian Models are widely used in target tracking and signal processing, since inference for these models are tractible (we only need to manipulate Gaussians).

## Example: 1-D Linear Gaussian Models

The following is a 1-dimensional example:

$$
\begin{align}
X_0 &\sim \mathcal{N}(0,1),\\
X_{n+1} &\sim\mathcal{N}(1.01 X_{n} + 0.2, 1), \\
Y_n &\sim \mathcal{N}(X_n, 10).
\end{align}
$$

Alternatively, we can write the above as:

$$
\begin{align}
&X_0 \sim \mathcal{N}(0,1), &\\
&X_{n+1} = 1.01 X_{n} + 0.2 + \xi, &\xi \sim \mathcal{N}(0, 1), \\
&Y_n = X_n + \sqrt{10} * \epsilon, &\epsilon \sim \mathcal{N}(0, 1).
\end{align}
$$



## Example: Simulating Data from a 1-D Linear Gaussian Model

```python
def LGM_simulate(a, b, c, d, e, f, sample_x, steps=100):
    x_current = sample_x(1)
    latent_vars = [x_current]
    observations = []
    
    for _ in range(steps):
        y_current = d * x_current + e + f * np.random.normal(0, 1)
        x_next = a * x_current + b  + c * np.random.normal(0, 1)
        observations.append(y_current)
        latent_vars.append(x_next)
        x_current = x_next
        
    return np.array(latent_vars[:-1]), np.array(observations)
```

```python
###define the parameters of the Linear Gaussian Model

#parameters for the underlying markov process: x_n+1 = a * x_n + b + c * N(0, 1)
a = 1.01
b = 0.2
c = 1.

#parameters for the observation model: y_n = d * x_n + e + f * N(0, 1)
d = 1.
e = 0.
f = np.sqrt(10)

#number of time steps to simulate
steps = 200

#initial distribution over latent state space
sample_x = lambda n: np.random.normal(0, 1, size=n)

def plot_lgm(ax):
    ###plot a single simulation: compare latent with observed

    #simulate the LGM once, for both latent and observed variables
    latent_vars, observations = LGM_simulate(a, b, c, d, e, f, sample_x, steps=steps)
    #plot
    ax[0].plot(np.arange(len(latent_vars)), latent_vars, color='red', label='latent states')
    ax[0].scatter(np.arange(len(observations)), observations, color='blue', label='observations', alpha=0.1)
    ax[0].legend(loc='best')
    ax[0].set_xlabel('Time Index')
    ax[0].set_title('Single Simulation of the LGM')

    ###plot multiple simulations: compare observed with the mean of observations
    simulations = 50 #number of simulations
    observation_mean = 0 #array of mean observed value for each time index

    for i in range(simulations):
        _, observations = LGM_simulate(a, b, c, d, e, f, sample_x, steps=steps)
        observation_mean += observations
        if i == 0:
            ax[1].plot(np.arange(len(observations)), observations, color='blue', label='observations', alpha=0.1)
        else:
            ax[1].plot(np.arange(len(observations)), observations, color='blue', alpha=0.1)

    observation_mean /= (simulations * 1.)

    ax[1].plot(np.arange(len(observations)), observation_mean, color='black', label='observation mean', alpha=1.)
    ax[1].legend(loc='best')
    ax[1].set_xlabel('Time Index')
    ax[1].set_title('Multiple Simulation of the LGM')
    
    return ax
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax = plot_lgm(ax)
plt.show()
```

## Inference
There are a number of inference problems associated with HMMs: <br>
**1.** learning the dynamics of the state or observation model <br>
**2.** estimating the probability distribution over one or more of the latent variables, $\{X_n\}_{n\geq 1}$, given a sequence of observations, $\{Y_n\}_{n\geq 1}$.

The task of estimating the latent variables can be further broken down:

**1. Filtering** computing the distribution of the last latent variable in the sequence given all observations, $p(X_n| Y_{1:n})$.<br><br>
**2. Smoothing** computing the distribution of a hidden latent variable in a past time step given all observations, $p(X_t| Y_{1:n})$ where $t < n$.<br><br>
**3. Most probable explanation** computing the joint distribution of all latent variables given all observations, $p(X_{1:n}| Y_{1:n})$. Or, alternatively, compute the most likely sequence of latent variable values given all observations:
$$
x^*_{1:n} = \underset{x_{1:n}}{\mathrm{argmax}}\; p(x_{1:n} | y_{1:n})
$$
We focus first on ***filtering***.

We make the following **simplifying assumptions**:
0. we know the model parameters, i.e. we know the parameters in the distribution $p(X_0)$, $p(X_{n+1} | X_{n})$ and $p(Y_n | X_n)$.
1. the underlying Markov process is homogeneous, that is, the transition probability density function is stationary. That is, for all $n\geq 0$, we have
$$
p(Y_n | X_n) = p(Y_{n+1} | X_{n+1}).
$$
2. given $\{X_n\}_{n\geq 0}$, the observations $\{Y_n\}_{n\geq 0}$ are independent.

# Kalman Filters

## Filtering: Infering $p(X_n| Y_{1:n})$

By Baye's Rule, the posterior marginal distribution $p(X_n| Y_{1:n})$ is given by
$$
p(X_n| Y_{1:n}) = \frac{p(Y_n | X_n) p(X_n | Y_{1:n-1})}{p(Y_n | Y_{1:n-1})}.
$$
Note that in the above $p(Y_n | X_n)$ is known; it is just the likelihood (distribution of the observed value given the latent state). 

Given $p(X_{n-1} | Y_{1:n-1})$, we can compute the unknown term $p(X_n | Y_{1:n-1})$ in the numerator:
$$
p(X_n | Y_{1:n-1}) = \int p(X_n | X_{n-1}) p(X_{n-1} | Y_{1:n-1}) dX_{n-1}. 
$$
Given $p(X_{n-1} | Y_{1:n-1})$, we can also compute the denominator $p(Y_n | Y_{1:n-1})$:
$$
p(Y_n | Y_{1:n-1}) = \int p(X_{n-1} | Y_{1:n-1}) p(X_n | X_{n-1}) p(Y_n | X_n) dX_{n-1:n}.
$$
Note that $p(X_n | X_{n-1})$ is known, it is the transition between latent states.

## Inducitive Algorithm for Inference

We've seen that given $p(X_{n-1} | Y_{1:n-1})$, we can compute $p(X_n| Y_{1:n}) = \frac{p(Y_n | X_n) p(X_n | Y_{1:n-1})}{p(Y_n | Y_{1:n-1})}$. This leads to the following inducitive algorithm for infering $p(X_n| Y_{1:n})$: 

0. (**Inductive Hypothesis**) suppose we have $p(X_{n-1} | Y_{1:n-1})$
1. (**Prediction step**) compute $p(X_n | Y_{1:n-1})$ 
2. (**Update step**) compute $p(X_n| Y_{1:n})$

The problem is that save for in very few cases, the integrals we need to compute in both the prediction and update steps are intractable! 

In the case of linear Gaussian models, the distributions in the prediction and update steps are Gaussian and can be computed in closed form, the resulting iterative filtering algorithm is known as the ***Kalman Filter***.

## Kalman Filters for 1-D Linear Guassian Models

Given a 1-D linear Gaussian model:
\begin{align}
&X_0 \sim \mathcal{N}(0, \Sigma)\\
&X_{n+1} = aX_n + b + c\xi,\;\; \xi\sim \mathcal{N}(0, 1) \quad \mathbf{(State\;Model)}\\
&Y_{n+1} = dX_{n+1} + e + f\epsilon,\;\; \epsilon\sim \mathcal{N}(0, 1) \quad \mathbf{(Observation\;Model)}
\end{align}

**0.** (**Inductive Hypothesis**) suppose we have $p(X_{n-1} | Y_{1:n-1}) = \mathcal{N}(\hat{x}_{n-1}, \hat{\sigma}^2_{n-1})$<br><br>
**1.** (**Prediction step**) compute 
$$p(X_n | Y_{1:n-1}) = \int p(X_n|X_{n-1})p(X_{n-1}|Y_{1:n-1})dX_{n-1} = \mathcal{N}(a\hat{x}_{n-1} + b, r_{n}),$$
where $r_n = a^2 \hat{\sigma}^2_{n-1} + c^2$.<br><br>
**2.** (**Update step**) compute $p(X_n| Y_{1:n}) = \frac{p(Y_t|X_t)p(X_t|Y_{1:t-1})}{p(Y_n | Y_{1:n-1})} = \mathcal{N}(\hat{x}_n, \hat{\sigma}_n)$, where 
\begin{align}
\hat{x}_n &= \hat{x}_{n-1} + K_n(y_n - d(a\hat{x}_{n-1} + b) + e ),\\
\hat{\sigma}_n &= r_n  (1 - K_n  d),\\
K_n &= \frac{r_n  d}{f^2 + d^2  r_n}.
\end{align}


## Example: Kalman Filtering for a 1-D Linear Gaussian Model
Recall our 1-dimensional linear Gaussian model example:
$$
\begin{align}
&X_0 \sim \mathcal{N}(0,1), &\\
&X_{n+1} = 1.01 X_{n} + 0.2 + \xi, &\xi \sim \mathcal{N}(0, 1), \\
&Y_n = X_n + \sqrt{10} \epsilon, &\epsilon \sim \mathcal{N}(0, 1).
\end{align}
$$

```python
def kalman_filter(y, a, b, c, d, e, f, mu_init=0, sigma_init=1.):
    steps = len(y)
    x_mean_current = mu_init
    x_sigma_current = sigma_init
    x_mus = [mu_init]
    x_sigmas = [sigma_init]

    for n in range(1, steps):
        #predict step
        mu = a * x_mean_current + b 
        r = a**2 * x_sigma_current + c**2
        y_current = d * mu + e
        
        #update step
        K = r * d / (f**2 + d**2 * r)
        x_mean_next = mu + K * (y[n] - y_current)
        x_sigma_next = r * (1 - K * d)
        
        #prep for next iteration
        x_mus.append(x_mean_next[0])
        x_sigmas.append(x_sigma_next)
        
        x_mean_current = x_mean_next
        x_sigma_current = x_sigma_next
        
    return np.array(x_mus), np.array(x_sigmas)
```

```python
###define the parameters of the Linear Gaussian Model

#parameters for the underlying markov process: x_n+1 = a * x_n + b + c * N(0, 1)
a = 1.01
b = 0.2
c = 1.

#parameters for the observation model: y_n = d * x_n + e + f * N(0, 1)
d = 1.
e = 0.
f = np.sqrt(10)

#number of time steps to simulate
steps = 60

#initial distribution over latent state space
sample_x = lambda n: np.random.normal(0, 1, size=n)

###generate data by simulating the LGM
latent_vars, observations = LGM_simulate(a, b, c, d, e, f, sample_x, steps=steps)
```

```python
###estimate latent states given observations using Kalman filtering
x_mus, x_sigmas = kalman_filter(observations, a, b, c, d, e, f, mu_init=0, sigma_init=10)
```

```python
def plot_lgm_filter(ax):
    ###compare true latent states with estimated latent states from Kalman filtering

    #plot ground truth
    ax.plot(np.arange(steps), latent_vars, 
            color='red', label='latent states')
    ax.scatter(np.arange(len(observations)), observations,
               color='black', s=5., alpha=0.3, label='observations')

    #plot estimated mean of marginal posterior over latent states
    ax.plot(np.arange(steps), x_mus, 
            color='blue', label='Kalman estimate mean over latent states')

    #plot estimated variance of marginal posterior over latent states
    ax.fill_between(np.arange(steps), x_mus + 2 * x_sigmas, x_mus - 2*x_sigmas, 
                    color='blue', alpha=0.1, 
                    label='Kalman estimate variance over latent states, +/- 2 std')

    ax.legend(loc='best')
    ax.set_xlabel('Time Index')
    ax.set_title('Kalman Filter Estimate of LGM States')
    
    return ax
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax = plot_lgm_filter(ax)
plt.show()
```

## Limitations of the Kalman Filter

The main limitation of the Kalman filter is that it assumes linear dynamics in the underlying HMM model, whereas many stochastic systems in real-life applications have non-linear dynamics.

For example, a ***stochastic volatility model (SVM)***, the transition is linear Gaussian while the observation is non-linear in the following way:

$$
\begin{align}
&X_0 \sim \mathcal{N}\left(0, \frac{\sigma^2}{1 - \alpha^2} \right)&\\
&X_{n+1} = \alpha \cdot X_{n} + \sigma\cdot\xi, &\xi \sim \mathcal{N}(0, 1)\\
&Y_{n+1} = \beta\cdot\epsilon\cdot\mathrm{exp}\left( \frac{X_{n+1}}{2}\right), & \epsilon \sim \mathcal{N}(0, 1)
\end{align}
$$

Alternatively, one can write the SVM model as:
$$
\begin{align}
X_0 &\sim \mathcal{N}\left(0, \frac{\sigma^2}{1 - \alpha^2} \right)\\
X_{n+1} | X_n &\sim \mathcal{N}(\alpha \cdot X_{n}, \sigma^2)\\
Y_{n+1} | X_{n+1} &\sim \mathcal{N}(0, \beta^2 \cdot \mathrm{exp}(X_{n+1}))
\end{align}
$$

Stochastic volatility models are often used in economics and finance to infer and predict underlying volatility from observed price or rate data.

## An Example of Non-linear Dynamics: Stochastic Volitility Model

In a stochastic volitility model,
$$
\begin{align}
X_0 &\sim \mathcal{N}\left(0, \frac{\sigma^2}{1 - \alpha^2} \right)\\
X_{n+1} | X_n &\sim \mathcal{N}(\alpha \cdot X_{n}, \sigma^2)\\
Y_{n+1} | X_{n+1} &\sim \mathcal{N}(0, \beta^2 \cdot \mathrm{exp}(X_{n+1}))
\end{align}
$$

$\mathrm{E}(Y_{n+1}/Y_n)$ is the expected return rate and the latent states represent the underlying volatility. High volatility means that the prices exhibit large random fluctuations, while high return rate means that the value of the prices increases rapidly on average.

```python
def SVM_simulate(sigma, alpha, beta, sample_x, steps=100):
    x_current = sample_x(1)
    latent_vars = [x_current]
    observations = []
    
    for _ in range(steps):
        y_current = beta * np.exp(x_current / 2.) * np.random.normal(0, 1) 
        x_next = alpha * x_current + sigma * np.random.normal(0, 1)
        observations.append(y_current)
        latent_vars.append(x_next)
        x_current = x_next
        
    return np.array(latent_vars[:-1]), np.array(observations)
```

```python
###define the parameters of the Stochastic Volatility Model

#parameters for the underlying markov process: x_n+1 = alpha * x_n + sigma * N(0, 1)
alpha = 0.91
sigma = 1.

#parameters for the observation model: y_n = beta * exp(x_n/2) * N(0, 1)
beta = 0.5

#number of time steps to simulate
steps = 500

#initial distribution over latent state space
sample_x = lambda n: np.random.normal(0, sigma**2 / (1. - alpha**2), size=n)
```

```python
def plot_svm(ax):
    ###plot a single simulation: compare latent with observed

    #simulate the SVM once, for both latent and observed variables
    latent_vars, observations = SVM_simulate(sigma, alpha, beta, sample_x, steps=steps)
    #plot
    ax[0].plot(np.arange(len(latent_vars)), latent_vars, color='red', linewidth=0.8, label='latent states')
    ax[0].scatter(np.arange(len(observations)), observations, color='blue', s=5., label='observations', alpha=0.2)
    ax[0].legend(loc='best')
    ax[0].set_ylim([-12, 12])
    ax[0].set_xlabel('Time Index')
    ax[0].set_title('Single Simulation of the SVM')

    ###plot multiple simulations: compare observed with the mean of observations
    simulations = 50 #number of simulations
    observation_mean = 0 #array of mean observed value for each time index

    for i in range(simulations):
        _, observations = SVM_simulate(sigma, alpha, beta, sample_x, steps=steps)
        observation_mean += observations
        if i == 0:
            ax[1].plot(np.arange(len(observations)), observations, color='blue', label='observations', alpha=0.1)
        else:
            ax[1].plot(np.arange(len(observations)), observations, color='blue', alpha=0.1)

    observation_mean /= (simulations * 1.)

    ax[1].plot(np.arange(len(observations)), observation_mean, color='black', label='observation mean', alpha=1.)
    ax[1].legend(loc='best')
    ax[1].set_ylim([-20, 20])
    ax[1].set_xlabel('Time Index')
    ax[1].set_title('Multiple Simulation of the SVM')
    
    return ax
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax = plot_svm(ax)
plt.show()
```
